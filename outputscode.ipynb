```python
import nbformat as nbf
import os
from nbconvert.preprocessors import ExecutePreprocessor
from nbconvert import HTMLExporter
import json
# Parameters
strategies = ['0','1'] #1 for Padding, 0 for Differential Privacy (FPA)
pads = ['100', '300', '350', '500', '700', '900', '1000', '1500'] #level 100, level 300, level 350, level 500, level 700, level 900, random, MTU
vectors = ['50', '40', '30', '25', '20', '15', '10', '5'] #Vector Size: 5, 10, 15, 20, 25, 30, 40, 50
ks = ['10', '20'] #coefficients: 10, 20 
epsilons = ['0.05', '0.5', '5', '10'] #epsilon: 0.05, 0.5, 5, 10


# Path to your notebook
notebook_path = r"C:\Users\user\Desktop\PROJE\maincode.ipynb"

# Load the notebook
with open(notebook_path, encoding='utf-8') as f:
    nb = nbf.read(f, as_version=4)

ep = ExecutePreprocessor(kernel_name='python3')

# Output directory
output_dir = r"C:\Users\user\Desktop\PROJE\Outputs" #output path
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate over all combinations of parameters
for strategy in strategies:
    if strategy == '1':
        # Iterate over pads and vectors
        for pad in pads:
            for vector in vectors:           
                # Update the second cell with the new parameters
                nb.cells[0].source = f"""\
                strategy= {strategy} 
                pad= {pad} 
                vector={vector}
                """
                # Execute the notebook
                ep.preprocess(nb, {'metadata': {'path': 'C:\\Users\\user\\Desktop\\PROJE'}})#code path

                # Export to HTML
                html_exporter = HTMLExporter()
                html_data, _ = html_exporter.from_notebook_node(nb)

                # Write the HTML data to a file
                output_filename = f"output_{strategy}_{pad}_{vector}.html"
                output_path = os.path.join(output_dir, output_filename)
                with open(output_path, 'w', encoding='utf-8') as f:
                    f.write(html_data)

    elif strategy == '0':
        # Iterate over vectors, ks, and epsilons
        for vector in vectors:
            for k in ks:
                for epsilon in epsilons:               
                    # Update the second cell with the new parameters
                    nb.cells[0].source = f"""\
                    strategy= {strategy} 
                    vector={vector} 
                    k={k} 
                    epsilon={epsilon}
                    """
                    # Execute the notebook
                    ep.preprocess(nb, {'metadata': {'path': 'C:\\Users\\user\\Desktop\\PROJE'}})#code path

                    # Export to HTML
                    html_exporter = HTMLExporter()
                    html_data, _ = html_exporter.from_notebook_node(nb)

                    # Write the HTML data to a file
                    output_filename = f"output_{strategy}_{vector}_{k}_{epsilon}.html"
                    output_path = os.path.join(output_dir, output_filename)
                    with open(output_path, 'w', encoding='utf-8') as f:
                        f.write(html_data)
```
